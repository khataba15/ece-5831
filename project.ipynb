```python
import pandas as pd
import numpy as np
from two_layer_net import TwoLayerNetwork
import matplotlib.pyplot as plt
import prepareDataset as p
```

```python
(train_data, train_label, test_data, test_label) = p.prepare_dataset()
```

```python
trained_model_filename = 'khan_project.pkl'
iterations = 5000
train_size = train_data.shape[0]
batch_size = 5
learning_rate = 0.06

batch_mask = np.random.choice(train_size, batch_size)

net = TwoLayerNetwork(input_size= 2*10, hidden_size=30, output_size=3)

net.fit(iterations, train_data, train_label, test_data, test_label, batch_size, learning_rate=learning_rate,backprop=True)


    
plt.figure()
x = np.arange(len(net.train_losses))
plt.plot(x, net.train_losses, label = 'train loss')
plt.xlabel('epochs')
plt.ylabel('loss')
plt.show()


plt.figure()
markers = {'train':'o', 'test': 's'}
x = np.arange(len(net.train_acc))
plt.plot(x, net.train_acc, label = 'train accuracy')
plt.plot(x, net.test_acc, label = 'test accuracy', linestyle = '--')
plt.xlabel('epochs')
plt.ylabel('accuracy')
plt.ylim(0,1.0)
plt.legend(loc='lower right')
plt.show()
```

```python
y = net.predict(test_data[9]) 
Actual_g = net.Gesture(np.argmax(y))
Expected_g = net.Gesture(np.argmax(test_label[9]))
print("The actual Gesture is {} and the expected Gesture is {}".format(Actual_g,Expected_g))
```
